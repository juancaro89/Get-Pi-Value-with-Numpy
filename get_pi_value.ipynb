```python
import numpy as np
import random 
import matplotlib.pyplot as plt
```

```python
inside_circle_count = 0
outside_circle_count = 0
```

```python
# generate random numbers between 0-1
for counter in range(0,10000):
    x = random.random()
    y = random.random()
    
    #the point is inside the circle?
    if(x**2 +y **2<1):
        
        #how many points will be inside the circle
        inside_circle_count= inside_circle_count +1
        plt.scatter(x,y)
    #the point is outside the circle
    elif(x**2 +y **2>1):
        #how many points will be outside the circle
        outside_circle_count = outside_circle_count +1

pi = 4*(inside_circle_count / (inside_circle_count +outside_circle_count))
print(pi)
plt.show()
    
```
